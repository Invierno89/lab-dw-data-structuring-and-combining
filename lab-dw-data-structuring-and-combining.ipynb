```python
import pandas as pd
```

```python
# Your code goes here

df_1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')

df_2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df_3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
```

```python
print(df_1.head())
print(df_2.head())
print(df_3.head())
```

```python
print(df_1.columns)
print(df_2.columns)
print(df_3.columns)
```

```python
df_1.columns = list(map(lambda title: title.lower(), df_1.columns))
df_2.columns = list(map(lambda title: title.lower(), df_2.columns))
df_3.columns = list(map(lambda title: title.lower(), df_3.columns))
print(df_1.columns)
print(df_2.columns)
print(df_3.columns)
```

```python
df_1.columns = list(map(lambda title: title.replace(" ", "_"), df_1.columns))
df_2.columns = list(map(lambda title: title.replace(" ", "_"), df_2.columns))
df_3.columns = list(map(lambda title: title.replace(" ", "_"), df_3.columns))
print(df_1.columns)
print(df_2.columns)
print(df_3.columns)
```

```python
df_1.rename(columns= {'st' : 'state'}, inplace = True)
df_2.rename(columns= {'st' : 'state'}, inplace = True)
print(df_1.columns)
print(df_2.columns)
print(df_3.columns)
```

```python
col = list(df_1.columns)
df_1 = df_1[col]
df_2 = df_2[col]
df_3 = df_3[col]
print(df_1.columns)
print(df_2.columns)
print(df_3.columns)
```

```python
df = pd.concat([df_1, df_2, df_3], axis=0)
df
```

```python
unknown_gender = 'Unknown'
df['gender'] = df['gender'].map({'Male' : 'M', 'Femal' : 'F', 'female' : 'F'}).fillna(unknown_gender)
```

```python
df['gender'].unique()
```

```python
df['state'] = df['state'].map({'AZ' : 'Arizona', 'WA' : 'Washington', 'Cali' : 'California', 'Nevada' : 'Nevada', 'Oregon' : 'Oregon'})
```

```python
df['state'].unique()
```

```python
df['education'] = df['education'].map({'Bachelors' : 'Bachelor', 'Master' : 'Master', 'High School or Below': 'High School or Below', 'College' : 'College', 'Doctor' : 'Doctor'})
```

```python
df['education'].unique()
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].apply(lambda x: str(x) if isinstance(x, float) else x, list(df['customer_lifetime_value']))

df['customer_lifetime_value'] = df['customer_lifetime_value'].apply(lambda x: x.replace("%", ""), list(df['customer_lifetime_value']))
```

```python
df['vehicle_class'] = df['vehicle_class'].map({'Four-Door Car' : 'Four-Door Car', 'Two-Door Car' : 'Two-Door Car', 'SUV' : 'SUV', 'Luxury SUV' : 'Luxury', 'Sports Car' : 'Luxury', 'Luxury Car' : 'Luxury'})
```

```python
df['vehicle_class'].unique()
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].apply(lambda x: isinstance(x, float) if type(x) == 'str' else x)
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(float).round(2)
```

```python
df['customer_lifetime_value'].unique()
```

```python
df['number_of_open_complaints'] = df['number_of_open_complaints'].apply(lambda x: x.split('/')[1] if isinstance(x, str) and '/' in x else x)


df['number_of_open_complaints'] = df['number_of_open_complaints'].astype(str)
```

```python
df['number_of_open_complaints'].unique()
```

```python
df['total_claim_amount'] = df['total_claim_amount'].round(2)
```

```python
df.tail(10)
```

```python
import numpy as np

df['customer_lifetime_value'] = df['customer_lifetime_value'].replace('nan', np.nan)
df['number_of_open_complaints'] = df['number_of_open_complaints'].replace('nan', np.nan)
```

```python
num_rows_null = df.isnull().all(axis=1).sum()

num_rows_null
```

```python
df = df.dropna(how='all')
```

```python
df['state'] = df['state'].replace(np.nan, 'Unknown')
df['gender'] = df['gender'].replace(np.nan, 'Unknown')
df['education'] = df['education'].replace(np.nan, 'Unknown')
```

```python
sum(df.duplicated())
```

```python
df = df.drop_duplicates()
sum(df.duplicated())
```

```python
df = df.reset_index(drop=True)
df
```

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Your code goes here
# df_marketing_pivot03 = df_marketing.pivot_table(index='policy_type', columns='month', values=[df_marketing['month'].value_counts()], aggfunc='sum')

complaints_by_policy_month = df_marketing.groupby(['policy_type', 'month']).agg({'number_of_open_complaints':'count'}).reset_index()
long_format_complaints = complaints_by_policy_month.melt(id_vars=['policy_type'], var_name='month', value_name='complaints')
```
